```python

X = data
y = data['Class']
```

```python
X.shape, y.shape
```

Now letâ€™s split the data into 2 parts : Training and Testing.

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score as ras
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
models = [LogisticRegression(), XGBClassifier(),
		SVC(kernel='rbf', probability=True),
		RandomForestClassifier(n_estimators=7,
								criterion='entropy', 
								random_state=7)]

for i in range(len(models)):
	models[i].fit(X_train, y_train)
	print(f'{models[i]} : ')
	
	train_preds = models[i].predict_proba(X_train)[:, 1]
	print('Training Accuracy : ', ras(y_train, train_preds))
	
	y_preds = models[i].predict_proba(X_test)[:, 1]
	print('Validation Accuracy : ', ras(y_test, y_preds))
	print()
```

```python
import sklearn
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

predictions = models[1].predict(X_test)
#predictions=models[i].predict_proba(X_test)[:, 1]
cm = confusion_matrix(y_test, predictions, labels=models[1].classes_) 
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=models[1].classes_)

#ConfusionMatrixDisplay(models[1], X_test, y_test)
disp.plot()
#plt.show()
```
